```julia
#### Power Method

function power(A, x, eps)

    lambda = 0
    e = 1.0
    t = 0
    
    while (e > eps && t < 1000)
        t = t + 1
        y = A*x
        x = y/norm(y)
        lambda1 = lambda
        lambda = x'*y

        # Erro
        e = abs(lambda1 - lambda)
    end
    return lambda
end
```

```julia
power(A, x, 1e-8)
```

```julia
A = randn(21, 21)
x = randn(21) + randn(21)*im
x = (randn(21) + randn(21)*im)/norm(x)
```

```julia
##### Inverse Iteration

function invpower(A, sigma, x, eps)
    
    lambda = 0.0
    e = 1.0
    t = 0
    
    while (e > eps && t < 1000)
        t = t + 1
        y = (A - sigma*I)\x
        x = y/norm(y)
        lambda1 = lambda
        lambda = x'*A*x

        # Erro
        e = abs(lambda1 - lambda)
    end
    return lambda
end
```

```julia
invpower(A, 1.35, x, 1e-5)
```

```julia
#### Comparando valores
sort(abs.(eigen(A).values))
```

```julia
# Orthogonal Iteration

function orthogonal(A, k, eps)
    
    (m, n) = size(A)
    x = randn(m, k)
    F = qr(x)
    Z = F.Q[:, 1:k]
    lambda = zeros(k)
    e = 1.0
    t = 0
    
    while (e > eps && t < 1000)
        t = t + 1
        F = qr(A*Z)
        Z = F.Q[:, 1:k]
        E = Z'*A*Z
        lambda1 = lambda
        lambda = eigen(E).values
        # Erro
        e = norm(lambda1 - lambda)
    end
    return lambda
end
```

```julia
orthogonal(A, 2, 1e-8)
```

```julia
function qriteration(A, eps)
    
    e = 1.0
    t = 0
    
    while (e > eps && t < 10000)
        t = t + 1
        A0 = A
        A = qr(A)
        A = A.R*A.Q

        # Erro
        e = norm(A - A0)
    end
    return A
end
```

```julia
A = randn(21, 21)
qriteration(A, 1e-8)
```

```julia
function qrshiftite(A, sigma, eps)
    
    e = 1.0
    t = 0
    
    while (e > eps && t < 50000)
        t = t + 1
        F = qr(A - sigma*I)
        A0 = A
        A = F.R*F.Q + sigma*I

        # Erro
        e = norm(A - A0)
    end
    return A
end
```

```julia
R = qrshiftite(A, 1.0, 1e-10)
```
