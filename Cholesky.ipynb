```julia
using(LinearAlgebra)
using(Plots)
```

```julia
##### Nível 2

function  Cholesky_factorize(A)

    L = zero(A)

    for j in 1:size(A, 1)

        L[j, j] = (A[j, j] - sum(L[j, 1:j-1].^2))^(.5)

        for i in (j+1):size(A, 1)
        
            L[i, j] = (A[i, j] - dot(L[i, 1:j-1], L[j, 1:j-1]))/L[j, j]

        end

    end

    return L
end
```

```julia
#### Nível 3 ######
function Cholesky_by_blocks(A, block_size)

    L = zero(A)
    C = copy(A)
    n = size(A, 1)


    for j in 1:block_size:n

        b = min(n - j + 1, block_size)

        C[j:(j+b-1), j:(j+b-1)] = Cholesky_factorize(C[j:(j + b - 1), j:(j + b - 1)])
        C[(j+b):n, j:( j + b - 1)] = C[(j+b):n, j:(j + b -1)]*inv(C[j:(j + b - 1), j:(j + b - 1)]')
        C[(j+b):n, (j + b):n] = C[(j+b):n, (j+b):n] - C[(j+b):n, j:(j + b - 1)]*C[(j+b):n, j:(j + b - 1)]'

    end

    return tril(C)
end
```

```julia
# Testando com Cholesky nível 2

n = 25
X = randn(n, n)

A = X*X'

L = Cholesky_factorize(A)

display(A)

display(L)

display(L*L')

display(A - L*L')

display(norm(A - L*L'))
```

```julia
n = 400
X = randn(n, n)

A = X*X'


@time Cholesky_factorize(A)
@time Cholesky_by_blocks(A, 20)
@time cholesky(A)
```

```julia
L = Cholesky_factorize(A)
C = Cholesky_by_blocks(A, 20)
F = cholesky(A)

display(norm(A - L*L'))
display(norm(A - C*C'))
display(norm(A - F.L*F.L'))
```
