```julia
using LinearAlgebra
using Printf
```

```julia
function j2c(A::Matrix{Float64},nome::String="A", ldA::Int=0)
    (m,n)=size(A)
    if ldA==0
        ldA=m
    end
    if (ldA<m)
        error("ERRO: ldA<m")
    end
    
    @printf("/* Matriz de %d linhas por %d colunas, formato Fortran (column first) */\n", m, n)
    @printf("const int ld%s=%d;\n",nome,ldA)    
    @printf("double %s[%d]={\n",nome,ldA*n)
    for j=1:n
        @printf("/* Coluna %d */\n",j)
        for i=1:ldA
            if i<=m
               @printf("%+.16e",A[i,j])
            else
                @printf("0.0")
            end
            if i<m || j<n
                @printf(",\n")
            else
                @printf("};\n")
            end
        end
    end
    @printf("\n")
end
```

```julia
A = randn(100, 12)
```

```julia
j2c(A)
```

```julia
b = randn(100, 1)
```

```julia
j2c(b, "b")
```

```julia
x = A\b
```

```julia
j2c(x, "x")
```
